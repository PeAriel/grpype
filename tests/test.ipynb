```python
import numpy as np
import matplotlib.pyplot as plt

from grpype.templates.spectral_models import band_function

%matplotlib inline
```

```python
x = 10**np.linspace(np.log10(10), np.log10(40000), 1000)
alphas = np.linspace(-1.5, 5, 5)
```

```python
for alpha in alphas:
    plt.plot(x, x**2*band_function([alpha, -3, 300], x), label=f"$\\alpha = {alpha}$")

# plt.plot(x, x**2*band_function([-1.5, -3, 2000], x), '--k')

plt.yscale("log")
plt.xscale("log")
plt.legend()
plt.show()
```
